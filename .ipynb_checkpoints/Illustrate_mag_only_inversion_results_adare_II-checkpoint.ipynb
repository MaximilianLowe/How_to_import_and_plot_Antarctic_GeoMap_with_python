```python
%%time

data_subset_ix = []

x_ix = xq.flatten()
y_ix = yq.flatten()

for i in range(len(Depth_sus)):
    z_ix = np.zeros(len(x_ix))+Depth_sus[i]
    inv_sus_depth_slice = inv_sus[i,:,:]
    inv_sus_ix = inv_sus_depth_slice.flatten()
    inv_data = np.vstack((x_ix , y_ix , z_ix , inv_sus_ix)).T
    
    for j in range(len(inv_data[:,0])):
        points_ix = Point(inv_data[j,0] , inv_data[j,1])
        if points_ix.within(polygon_geom) == True:
            data_within_polygon_ix = inv_data[j,:]
            data_subset_ix.append(data_within_polygon_ix)

data_subset = np.array(data_subset_ix) 
np.savetxt(inv_run+"inv_sus_hallit_xyzv.txt", data_subset , delimiter=',')
```
